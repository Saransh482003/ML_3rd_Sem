```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
coddist = lambda a,b : np.sqrt(((a-b)**2).sum()).round(3)
```

```python
clust_1 = np.array([[random.uniform(0,2),random.uniform(0,2)] for _ in range(100)])
clust_2 = np.array([[random.uniform(2,4),random.uniform(2,4)] for _ in range(100)])
```

```python
sns.set()
```

```python
plt.figure(figsize=(16,9))

sns.scatterplot(clust_1[:,0],clust_1[:,1], marker="P",s=200, color="#051e3e")
sns.scatterplot(clust_2[:,0],clust_2[:,1], marker="^",s=200, color="#851e3e")
```

```python
c_1 = np.array([clust_1[:,0].mean(),clust_1[:,1].mean()])
c_2 = np.array([clust_2[:,0].mean(),clust_2[:,1].mean()])
```

```python
plt.figure(figsize=(16,9))

sns.scatterplot(clust_1[:,0],clust_1[:,1], marker="P",s=200, color="#051e3e")
sns.scatterplot(clust_2[:,0],clust_2[:,1], marker="^",s=200, color="#851e3e")
sns.scatterplot([c_1[0]],[c_1[1]], marker = "o", s=200, color = "g")
sns.scatterplot([c_2[0]],[c_2[1]], marker = "o", s=200, color = "#fe8a71")
```

```python
x = np.array([float(i) for i in input().split(",")])
print(f"Belongs to Cluster {'Plus' if coddist(c_1,x)-coddist(c_2,x)<0 else 'Triange'}")

plt.figure(figsize=(16,9))

sns.scatterplot(clust_1[:,0],clust_1[:,1], marker="P",s=200, color="#051e3e")
sns.scatterplot(clust_2[:,0],clust_2[:,1], marker="^",s=200, color="#851e3e")
sns.scatterplot([c_1[0]],[c_1[1]], marker = "o", s=200, color = "g")
sns.scatterplot([c_2[0]],[c_2[1]], marker = "o", s=200, color = "#fe8a71")
sns.scatterplot([x[0]],[x[1]], marker = "D", s=200, color = "k")
plt.show()
```

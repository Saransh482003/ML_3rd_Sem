```python
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def make_blobs(samples,features,centers):
    fea=[]
    for i in centers:
        cen=[]
        for j in range(samples):
            sam=[]
            for k in range(features):
                sam.append(np.random.uniform(i[0],i[1]))
            cen.append(sam)
        fea.append(cen)
    return fea
```

```python
a = make_blobs(100,5,[[0,2],[2,4]])
```

```python
x,y = np.array(a[0]),np.array(a[1])
```

```python
x
```

```python
sns.set()
```

## Multi-dimentional Plotting

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x[:,0],x[:,1],x[:,2],x[:,3],x[:,4],legend=False)
sns.scatterplot(y[:,0],y[:,1],y[:,2],y[:,3],y[:,4],legend=False)
```

## Calculating Centers

```python
c_x = np.array([[x[:,i].mean()] for i in range(len(x[0,:]))])
c_y = np.array([[y[:,i].mean()] for i in range(len(y[0,:]))])
```

```python
c_x,c_y
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x[:,0],x[:,1],legend=False,marker=",",color="#900C3F",s=100)
sns.scatterplot(y[:,0],y[:,1],legend=False,marker="p",color="#FFC300",s=100)
sns.scatterplot(c_x[0],c_x[1],c_x[2],c_x[3],c_x[4],color="r",s=300,marker=",")
sns.scatterplot(c_y[0],c_y[1],c_y[2],c_y[3],c_y[4],color="g",s=300,marker="^")
```

## Cluster Allotment

```python
coddist = lambda a,b : np.sqrt(((a-b)**2).sum()).round(3)
```

```python
a = np.array([int(i) for i in input("Enter Coordinated : ").split(",")])
b = [coddist(i,a) for i in [c_x,c_y]]
print(f"Cluster {b.index(min(b))+1} Assigned to the new point.")
plt.figure(figsize=(16,9))
sns.scatterplot(x[:,0],x[:,1],legend=False,marker=",",color="#900C3F",s=100)
sns.scatterplot(y[:,0],y[:,1],legend=False,marker="p",color="#FFC300",s=100)
sns.scatterplot(c_x[0],c_x[1],c_x[2],c_x[3],c_x[4],color="r",s=300,marker=",")
sns.scatterplot(c_y[0],c_y[1],c_y[2],c_y[3],c_y[4],color="g",s=300,marker="^")
sns.scatterplot([a[0]],[a[1]],s=200,color="k",marker="P")
```
